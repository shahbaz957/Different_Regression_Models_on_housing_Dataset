```python
import sklearn as sk
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("C:/Users/Shahbaz Baig/Desktop/Data Science/USA_Housing.csv")
```

```python
df
```

```python
#five number theory
df.describe()
```

```python
#Although this data is already cleaned
df.dropna(inplace=True)
```

```python
#this is the textual column so i remove it for the sake of simplicty otherwise i have to do hot vector representation
df.drop('Address',inplace=True,axis=1)
```

```python
df
```

```python
x=df.drop('Price',axis=1)   #Input Data for Ml model
x
```

```python
y=df['Price'] #Output Data
```

```python
y
```

```python
from sklearn import preprocessing
pre=preprocessing.StandardScaler().fit(x)
x_transform=pre.fit_transform(x)
pre=preprocessing.StandardScaler().fit(y.to_frame())
y_transform=pre.fit_transform(y.to_frame())
```

```python
x_transform.shape
x_transform
```

```python
y_transform
```

```python
#for training of the model i divide the data in such a way that 80% data for training and 20% for testing the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_transform,y_transform,test_size=0.2,random_state=101)
```

```python
x_train.shape #ðŸ˜Š
```

```python
x_test.shape
```

```python
y_test.shape
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
l_reg=LinearRegression()
l_reg.fit(x_train,y_train)
```

```python
y_pred=l_reg.predict(x_test)
y_pred.shape
```

```python
sns.scatterplot(x=y_test.ravel(),y=y_pred.ravel(),color='blue',label='Actual Datapoints')
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color='red',label='Ideal Line')
plt.legend()
plt.show()
```

```python
print(y_test.shape)
print(y_pred.shape)
```

```python
result=np.column_stack((y_test,y_pred))
print('     Actual Values -- Model Values')
print('--'*30)
for a,p in result:
  print(f'{a:12.2f}  {p:12.2f}')
```

```python
#check the performance of my trained model
residual = a-y_pred.reshape(-1)
print(residual)
```

```python
sns.distplot(residual,kde=True)
```

# Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
print('MSE in Linear Regression :',mse)
```

```python
from sklearn.tree import DecisionTreeRegressor
Dtr=DecisionTreeRegressor()
Dtr.fit(x_train,y_train)
y_pred_Dt=Dtr.predict(x_test)
mse_Dt=mean_squared_error(y_test,y_pred_Dt)
print('MSE of Decision tree Regressor is :',mse_Dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
mse_rf=mean_squared_error(y_test,y_pred_rf)
print('MSE of Random Forest is :',mse_rf)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
y_pred_gb=gb.predict(x_test)
mse_gb=mean_squared_error(y_test,y_pred_gb)
print('MSE of Gradient Boosting is :',mse_gb)
```

```python
model_progress={
    'Linear Regression':0.081017255,
    'Decision Tree Regressor':0.256359949,
    'Gradient Boosting':0.099529316,
    'Random Forest':0.120646885
}
print('Final result and progress of model with Ranking')
print('--'*30)
sorted_score=sorted(model_progress.items(),key=lambda x:x[1]) #it will order the elements of the dictionary in ascending order according to the values of keys.
for rank,(name,score) in enumerate(sorted_score,start=1):
  print(f'{rank}. {name} -->  {score}')
```
